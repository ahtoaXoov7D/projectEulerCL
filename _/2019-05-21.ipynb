```python
def s(n):
    cards = list(range(n))
    count = 0
    while True:
        cards = sum(zip(cards[:n//2], cards[n//2:]), ())
        count += 1
        if cards[1] == 1:
            return count
        if count > 8:
            return -1
```

```python
print(s(86))
```

```python
for n in range(2, 100_000, 2):
    count = s(n)
    if count == 8:
        print(n, count) 
```

```python
def nxt(n, i):
    m = n//2
    if i <= m: 
        return 2*i - 1
    return 2 * (i - m
               ) 
```

```python
def fake(n):
    count = 0
    i = 2
    while True:
        i = nxt(n, i) 
        count += 1
        if i == 2:
            return count
        if count > 60: 
            return -1
```

```python
print("--", fake(52))
```

```python
sum_eyes = 0
for i in range(2, 10_000_000, 2):
    fk = fake(i)
    if fk == 60:
        print(i, fk)
        sum_eyes += i
        
print (sum_eyes)
```

```python
package main

import "fmt"

func outer(s_n, n int) int {

	half := n / 2
	count := 0
	nxt := 2

	for {
		if nxt <= half {
			nxt = (nxt + nxt) - 1
		} else {
			nxt = (nxt + nxt) - n
		}
		count++
		if nxt == 2 {
			return count
		}
		if count > s_n {
			return -1
		}
	}

	return count
}

func main() {
	for s_n := 2; s_n < 20; s_n += 1 {
		sum := 0
		prev := 0

		for i := 2; i < 1000000; i += 2 {
			count := outer(s_n, i)
			if count == s_n {
				fmt.Printf("%d -> %d (%d)\n", i, count, i-prev)
				sum += i
				prev = i
			}
		}

		fmt.Printf("The sum of i's for %d is %d\n", s_n, sum)
	}
}
```
